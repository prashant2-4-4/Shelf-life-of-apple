```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import keras
from keras.layers import Dense , Flatten  , Conv2D , MaxPool2D , Dropout
from PIL import Image

from tensorflow.keras.utils import to_categorical
```

```python
first_ = '../input/fruit-and-vegetable-dataset-for-shelf-life/Apple(1-5)'
second = '../input/fruit-and-vegetable-dataset-for-shelf-life/Apple(5-10)'
third = '../input/fruit-and-vegetable-dataset-for-shelf-life/Apple(10-14)'
```

```python
# main_dir = '../input/fruit-and-vegetable-dataset-for-shelf-life'
```

```python
import cv2
import os
all_images = []
predic_images = []
lst = [first_ , second , third]
k = -1
for j in lst:
    k += 1
    for filename in os.listdir(j):
        img = cv2.imread(os.path.join(j, filename))
        img = cv2.resize(img , (300,300))
        if img is not None:
            predic_images.append(k)
            img = np.array(img)
            all_images.append(img)
            
print(len(all_images)) 
    
```

```python
all_images = np.array(all_images)
predic_images = np.array(predic_images)
```

```python
import cv2
img = cv2.imread('../input/fruit-and-vegetable-dataset-for-shelf-life/Apple(10-14)/frame1.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)		# this converts it into RGB

plt.imshow(rgb_img)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    all_images, predic_images, test_size=0.25, random_state=0)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
y_test.shape
```

```python
y_train = to_categorical(y_train ,3)
y_test = to_categorical(y_test ,3 )
```

```python
y_test.shape
```

```python
y_test
```

```python
from keras.models import Sequential
from keras.layers import MaxPooling2D
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(300,300,3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python

histroy = model.fit(X_train , y_train , epochs = 20 , validation_data= (X_test , y_test))
```

* PLotting of loss vs accuracy graph using matplotlib

```python
plt.figure(0)
plt.plot(histroy.history['accuracy'] , label = 'Training accuracy')
plt.plot(histroy.history['val_accuracy'] , label = 'val accuracy')
plt.title('accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
```

```python
plt.figure(0)
plt.plot(histroy.history['loss'] , label = 'val training')
plt.plot(histroy.history['val_loss'] , label = 'val loss')
plt.title('loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
```
